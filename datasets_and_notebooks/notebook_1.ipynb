```python
import cv2
from deepface import DeepFace
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("neutral_face.jpeg")
```

```python
plt.imshow(img)
# It shows the BGR image
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
predictions= DeepFace.analyze(img_rgb, actions=[ 'emotion'])
```

```python
print(predictions)
```

```python
predictions[0]['dominant_emotion']
```

```python
faceCascade= cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = faceCascade.detectMultiScale(gray, 1.1, 4)

for x, y, w, h in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(img, predictions[0]['dominant_emotion'],(0,50),font,1,(0,0,255),2,cv2.LINE_4)
```
